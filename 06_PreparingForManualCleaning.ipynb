## Merging all steps

#### From the final merged table we would drop those that have cnt < 3 and top similarity score 70 (meaning that the next most similar name after cleaning step 7 is 70% similar, so even that is quite far, implying that this name is unique...)

```python
df = df_00.merge(df_01, on = 'step01', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_12, on = 'step1', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_23, on = 'step2', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_34, on = 'step3', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_45, on = 'step4', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_56, on = 'step5', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_7, on = 'step6', left_index=False, right_index=False )
df = df.drop_duplicates()

print(len(df))
df.head()
```

```python
fout = open('Ikon/cleaning_steps/NAMES_ALGO_CLEANED_MERGED.dat', 'w')
print(len(set(list(df.step6))))
for n in sorted(list(set(df.step6))):
    fout.write(n + '\n')
fout.close()
```
