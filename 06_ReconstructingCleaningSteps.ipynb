## Merging all steps

#### From the final merged table we would drop those that have cnt < 3 and top similarity score 70 (meaning that the next most similar name after cleaning step 7 is 70% similar, so even that is quite far, implying that this name is unique...)

```python
df = df_01.merge(df_12, on = 'step1', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_23, on = 'step2', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_34, on = 'step3', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_45, on = 'step4', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_56, on = 'step5', left_index=False, right_index=False )
df = df.drop_duplicates()

df = df.merge(df_7, on = 'step6', left_index=False, right_index=False )
df = df.drop_duplicates()

print(len(df))
df.head()
```

## All the steps and versions combined

#### We should filter based on the two version... parameters

```python
df.head()
```

```python
df_drop = df[(df['version 6 - cnt'] < 3) & (df['version 6 - top sim'] < 75)]

names_to_drop = list(df_drop.step6)

df_drop.head()
```

```python
df_for_manual = df[~df.step6.isin(names_to_drop)]
df_for_manual
```

## Graph components

```python
df_for_manual = df_for_manual.merge(df_8, on = 'step6', left_index=False, right_index=False )
df_for_manual = df_for_manual.drop_duplicates()
df_for_manual
```

```python
df_for_manual2 = df_for_manual[['similarity_component_id','step0', 'step6', 'version 6 - top sim', 'version 6 - cnt']]
df_for_manual2 = df_for_manual2.sort_values(by = 'similarity_component_id')
df_for_manual2 = df_for_manual2.rename(columns = {'step0' :  'original_form', 'step6' : 'cleaned_form', 'version 6 - cnt' : 'cleaned_frequency', 'version 6 - top sim' :  'similarity_score'})

df_for_manual2
```
